```python
# a) Parse each DNA information in iterative fashion and save it in Biopython record.
#sk-
#eNOV4Vu9Yi1UhmjpgUUwT3BlbkFJMIR3
#7FHkw3f6tfpS5PKj
from Bio import SeqIO
from Bio.SeqRecord import SeqRecord

file = 'ls_orchid.fasta'

records=[]

with open(file,'r') as fasta_file:
    for record in SeqIO.parse(fasta_file,"fasta"):
        #bioRecord = SeqRecord(seq=record.seq, id=record.id, name=record.name, description=record.description, 
        #                      annotations=record.annotations, features=record.features, letter_annotations=record.letter_annotations)
        records.append(record)
print(f'Number of records : {len(records)}')
```

```python
#b)	Align the DNA sequence in the record with sequences stored in any three online databases using BLAST tool.

from Bio.Blast import NCBIWWW

onlineDatabases=['nr','refseq_rna','pdb']
results=[]
for db in onlineDatabases:
    print(f'Checking for database : {db}')
    for record in records:
        result_handle = NCBIWWW.qblast("blastn",db,record.seq)
        results.append(result_handle.read())
        #with open(f"blast_results_{record.id}.xml", "w") as result_file:
        #    result_file.write(result_handle.read())
        result_handle.close()
    print('-----------------------------------------')
```

```python
#c)	Save each alignment in an XML file at the local drive.

from Bio.Blast import NCBIWWW
from Bio import SeqIO

# Path to the ls_orchid.fasta file
record = records[0]

# Perform BLAST search against three online databases
databases = ["nr", "refseq_rna", "pdb"]
for db in databases:
    print(f"Performing BLAST search against {db} database...")
    
    # Perform the BLAST search
    result_handle = NCBIWWW.qblast("blastn", db, record.seq)
    
    # Save the BLAST results to a file (optional)
    with open(f"blast_results_{db}.xml", "w") as result_file:
        result_file.write(result_handle.read())
    
    # Close the result handle
    result_handle.close()
    
    print(f"BLAST search against {db} completed.\n")
```

```python
#d)	Open the XML file and read the alignment record that contains all the alignments.
from Bio.Blast import NCBIXML

files = ['blast_results_nr.xml','blast_results_pdb.xml','blast_results_refseq_rna.xml']

for file in files:
    with open(file,'r') as result_file:
        print(file)
        blast_records = NCBIXML.parse(result_file)
        
        for blast_record in blast_records:
            for alignment in blast_record.alignments:
                print(f'Allignment title : {alignment.title}')
                
                for hsp in alignment.hsps:
                    print(f"  HSP Score: {hsp.score}")
                    print(f"  HSP E-value: {hsp.expect}")
                    print(f"  HSP Query: {hsp.query}")
                    print(f"  HSP Match: {hsp.match}")
                    print(f"  HSP Subject: {hsp.sbjct}")
                    print("\n") 
```

```python
# e)	Set a threshold value to pick the top 10 HSPs and print the following information:
# a.	Alignment title
# b.	Alignment length
# c.	Expected value
# d.	Query subsequence [75:150]
# e.	Match subsequence [75:150]
# f.	Target subsequence [75:150]

for file in files:
    with open(file,'r') as result_file:
        print(file)
        blast_records = NCBIXML.parse(result_file)
        
        for blast_record in blast_records:
            for alignment in blast_record.alignments:
                sorted_hsps = sorted(alignment.hsps, key=lambda x: x.score, reverse=True)
                
                topHSPS = sorted_hsps[:10]
                
                for i, hsp in enumerate(topHSPS):
                    print(f"\nTop HSP {i+1}:")
                    print(f"  Alignment Title: {blast_record.alignments[0].title}")
                    print(f"  Alignment Length: {hsp.align_length}")
                    print(f"  Expected Value: {hsp.expect}")
                
                    # Subsequences [75:150]
                    query_subsequence = hsp.query[74:150]
                    match_subsequence = hsp.match[74:150]
                    subject_subsequence = hsp.sbjct[74:150]

                    print(f"  Query Subsequence [75:150]: {query_subsequence}")
                    print(f"  Match Subsequence [75:150]: {match_subsequence}")
                    print(f"  Target Subsequence [75:150]: {subject_subsequence}")
```

```python
#a)

from Bio import SearchIO

# Specify the path to your XML file
xml_file_path = "./blast_results_nr.xml"

# Use the read function to parse the XML file into a QueryResult object
query_result = SearchIO.read(xml_file_path, "blast-xml")

```

```python
#c)

# Iterate over hits and print ID, sequence length, and number of HSPs
for hit in query_result:
    print(f"Hit ID: {hit.id}")
    print(f"Sequence Length: {hit.seq_len}")
    print(f"Number of HSPs: {len(hit.hsps)}")
    print()  # Add an empty line for better readability
```

```python
#d)

for hit_num, hit in enumerate(query_result.hits, start=1):
    print(f"\nHit {hit_num}:")
    print(f"Hit ID: {hit.id}")
    print(f"Description: {hit.description}")
    print(f"Sequence Length: {hit.seq_len}")
    print(f"Number of HSPs: {len(hit.hsps)}")
```

```python
#e)

# Iterate over hits, create Hit and HSP objects, and print information
for hit_num, hit in enumerate(query_result.hits, start=1):
    print(f"\nHit {hit_num}:")
    print(f"Hit ID: {hit.id}")
    print(f"Number of HSPs: {len(hit.hsps)}")

    # Iterate over HSPs and print information
    for hsp_num, hsp in enumerate(hit.hsps, start=1):
        print(f"\nHSP {hsp_num}:")
        print(f"E-value: {hsp.evalue}")
        print(f"Query Range: {hsp.query_range}")
        print(f"Hit Range: {hsp.hit_range}")
        print(f"Alignment Length: {hsp.aln_span}")
        print(f"Alignment: {hsp.aln}")
        print(f"Query Sequence: {hsp.query.seq}")
        print(f"Hit Sequence: {hsp.hit.seq}")
```
